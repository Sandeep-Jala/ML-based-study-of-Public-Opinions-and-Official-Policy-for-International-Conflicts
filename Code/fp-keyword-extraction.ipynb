```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from newspaper import Article
from wordcloud import WordCloud, STOPWORDS
```

```python
def extract_keywords(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        article.nlp()
        article_keywords = article.keywords
        return article_keywords
    except Exception as e:
        print(f"Error extracting keywords for {url}: {e}")
        return []
article_keywords = []
# urls_to_scrape = ['https://news.un.org/en/story/2023/10/1142347',
#                   'https://news.un.org/en/story/2023/10/1142792',
#                   'https://www.bbc.com/news/world-middle-east-67039975',
#                   'https://news.un.org/en/story/2023/10/1142742']

for url in urls_to_scrape:
    article_keywords.extend(extract_keywords(url))

keywords_df=pd.read_csv('/kaggle/input/final-project/keywords.csv')
keywords_df['flag']=keywords_df['flag'].fillna(1)
keywords=list(keywords_df[keywords_df['flag']==1]['keywords'])
# Convert the list of keywords to a string for WordCloud
text = ' '.join(article_keywords)
stopwords = set(STOPWORDS)

# Generate a word cloud based on word frequencies
word_freq = Counter(article_keywords)
wordcloud = WordCloud(background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=40, 
        colormap='viridis', collocations=False,
        scale=3).generate_from_frequencies(word_freq)

# Display the generated word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
